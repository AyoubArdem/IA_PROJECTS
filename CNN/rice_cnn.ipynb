```python
import random


images = []

for sub in os.listdir(DataRice):
  path = os.path.join(DataRice, sub)
  if os.path.isdir(path):
    for img in os.listdir(path):
      images.append([os.path.join(path, img), img.split()[0] ])

random.shuffle(images)

for i in range(10):
  print(images[i])
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(images, test_size=0.2)
```

```python

X_Train = []
Y_Train = []
X_Test = []
Y_Test = []

for feature, label in train:
  X_Train.append(feature)
  Y_Train.append(label)

for feature, label in test:
  X_Test.append(feature)
  Y_Test.append(label)

```

```python
import cv2

Test_data = []
Train_data = []

for i in X_Test[:2000]:
  img = cv2.imread(i)
  img_tr = cv2.resize(img,(128,128))
  Test_data.append(img_tr)

for i in X_Train[:9000]:
  img = cv2.imread(i)
  img_tr = cv2.resize(img,(128,128))
  Train_data.append(img_tr)
```

```python
Train_data = np.array(Train_data, dtype="float32") / 255.0
Test_data  = np.array(Test_data, dtype="float32") / 255.0

print(plt.imshow(Train_data[0]))
print(Train_data[0].shape)
```

```python
# Encoder les labels
labels = {'arborio':0, 'basmati':1, 'ipsala':2, 'jasmine':3, 'karacadag':4}
Y_Train = [labels[i.lower()] for i in Y_Train[:9000]]
Y_Test  = [labels[i.lower()] for i in Y_Test[:2000]]
```

```python
import tensorflow.keras as tf
from tensorflow.keras.utils import to_categorical

Y_Train = to_categorical(Y_Train, num_classes=5)
Y_Test = to_categorical(Y_Test, num_classes=5)
```

```python
import tensorflow as tf_lib
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Sequential

model = Sequential([
    Input(shape=(128,128,3)),
    Conv2D(32,(3,3),padding='same',activation='relu'),
    Conv2D(32,(3,3),activation='relu',padding='same'),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(64,(3,3),activation='relu',padding='same'),
    Conv2D(64,(3,3),activation='relu',padding='same'),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(128,(3,3),activation='relu',padding='same'),
    Conv2D(128,(3,3),activation='relu',padding='same'),
    MaxPooling2D(pool_size=(2,2)),

    Flatten(),
    Dense(128,activation='relu'),
    Dense(5,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(Train_data,Y_Train,epochs=10,batch_size=32,validation_data=(Test_data,Y_Test))
```

```python
model.summary()
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
sample = np.expand_dims(Test_data[1], axis=0)
pred = model.predict(sample)
res = np.argmax(pred)

print("Résultat prédiction:", res)

if res == 0:
    print("Arborio")
elif res == 1:
    print("Basmati")
elif res == 2:
    print("Ipsala")
elif res == 3:
    print("Jasmine")
else :
  print("Karcadag")

plt.imshow(cv2.imread(X_Test[1]))
```
